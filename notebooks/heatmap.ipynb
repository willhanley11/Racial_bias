```python
import folium
import pandas as pd
from folium.plugins import HeatMap
```

```python
# read in data
fatal = pd.read_csv('folium_data.csv')
```

```python
# Race value counts
fatal['Race'].value_counts()
```

```python
# list of longitudes and latitutudes of incidents
locations_list = fatal[['Latitude','Longitude']].values.tolist()

# create base layer of folium map on the USA
m = folium.Map(
    location=[40, -95],
    tiles='cartodbpositron',
    zoom_start=4)

# Heat map layer
hm = HeatMap(locations_list,
                     min_opacity=.3,
                     radius=10, blur=16,
                     max_zoom=10)


# add each point to base map
for point in range(0, round(len(locations_list)/25),2):
    folium.Map(locations_list[point]).add_to(m)

# add heatmap layer
m.add_child(hm)
```
